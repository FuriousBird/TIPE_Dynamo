```python
import numpy as np, matplotlib.pyplot as plt
```

```python
dat = np.loadtxt("mesureB_n.csv", delimiter=",", dtype=np.float64)
```

```python
sqdat = np.sqrt(np.abs(dat))*np.sign(dat)
absmax = np.max(np.abs(sqdat))
mapped_data = (sqdat+absmax)/(2*absmax)
```

```python
plt.suptitle("$B.u_y = f(x,y)$")
plt.title("Passé en racine carrée pour visualisation")

plt.xticks(ticks=np.arange(0, 10.5, 0.5))
plt.yticks(ticks=np.arange(0, 10.5, 0.5))

plt.gca().set_xticks(np.arange(0, mapped_data.shape[1], 1))
plt.gca().set_yticks(np.arange(0, mapped_data.shape[0], 1))
plt.gca().set_xticklabels(np.arange(0, 10.5, 0.5)-5, rotation=90)
plt.gca().set_yticklabels(np.arange(9.5, -0.5, -0.5) )
plt.imshow(mapped_data, cmap="plasma", vmin=0, vmax=1, interpolation="nearest")
```
