```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('reelight_dat.txt', delimiter='\t', dtype="float", skiprows=1)

#columns being: time	E	U(D2)	I(D2)	I(E)
time = data[:, 0]
E = data[:, 1]
U = data[:, 2]
I_D2 = data[:, 3]
I_E = data[:, 4]

PowE = I_E * E
PowD2 = I_D2 * U

fig, ax = plt.subplots(2, 1, figsize=(10, 8), sharex=True)
ax[0].plot(time, PowE, label='Power E', color='blue')
ax[0].plot(time, PowD2, label='Power D2', color='red')
ax[0].set_ylabel('Power (W)')
ax[0].legend()
ax[0].set_title('Power vs Time')
ax[1].plot(time, I_E, label='I(E)', color='blue')
ax[1].plot(time, I_D2, label='I(D2)', color='red')
ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('Current (A)')
ax[1].legend()
ax[1].set_title('Current vs Time')
plt.tight_layout()
plt.savefig('power_current_plot.png')
plt.show()
```

```python
#then calculate the average power input/output
avg_PowE = np.mean(PowE)
avg_PowD2 = np.mean(PowD2)
print(f'Average Power E: {avg_PowE:.2f} W')
print(f'Average Power D2: {avg_PowD2:.2f} W')
#power factor
power_factor = abs(avg_PowD2 / avg_PowE)
print(f'Power Factor: {power_factor:.2f}')
```
